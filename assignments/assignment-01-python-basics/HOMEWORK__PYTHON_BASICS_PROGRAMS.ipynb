## Program 17: Reverse a String

**Problem**: Read a string and print it reversed.

**Input**:
- One string `s`

**Output**:
- Reversed string

**Sample Input**:
```
Machine
```
**Sample Output**:
```
enihcaM
```


````python
name = list(input("Enter your name"))
start = 0
end = len(name) -1
while start <= end:
    temp = name[start]
    name[start] = name[end]
    name[end] = temp
    start +=1
    end-=1
print("".join(name))

````

## Program 18: Palindrome String

**Problem**: Read a string and print `Palindrome` or `Not Palindrome`.

**Input**:
- One string `s`

**Output**:
- `Palindrome` or `Not Palindrome`

**Sample Input**:
```
level
```
**Sample Output**:
```
Palindrome
```


````python
inputName = (input("Enter your name"))
name = list(inputName)
def reverseStr(name):
    start = 0
    end = len(name) -1
    while start <= end:
        temp = name[start]
        name[start] = name[end]
        name[end] = temp
        start +=1
        end-=1
    return ("".join(name))
reverseName = reverseStr(name)
if inputName == reverseName:
    print("Palindrome")
else:
    print("Not Palindrome")


````
