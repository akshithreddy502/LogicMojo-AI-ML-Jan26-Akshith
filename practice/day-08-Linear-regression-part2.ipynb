# OVERFITTING VS UNDERFITTING

# 1. The Goal of Machine Learning

When we train a model, we want it to:

Learn patterns from training data
AND
work well on new unseen data

If the model fails to do this, two problems happen:

Underfitting

Overfitting

# 2. Underfitting

Underfitting happens when the model is too simple and cannot learn the pattern in the data.

It performs poorly on both training data and test data.

Example

Suppose the real relationship is curved.

Actual pattern

y
|
|       *
|    *
|  *
| *
|*
+----------------
        x

But the model uses a straight line that does not fit the pattern.

Bad model (underfitting)

y

|       *
|    *
|  *
| *
|*
+----------------
  ---------
        x

The model misses the pattern.

# 3. Overfitting

Overfitting happens when the model learns the training data too perfectly, including noise.

It performs very well on training data but poorly on new data.

Example

Suppose data points look like this:

y
|
|     *
|   *
|      *
| *
|        *

        x

The model creates a very complex curve just to pass through every point.

Overfitting

# Why Overfitting Happens in Linear Regression

Overfitting usually happens when:

1️⃣ Too many features

Example:

Salary = f(

experience,

age,

city,

car,

shoe_size,

favorite_color

)

Many features may not matter.

2️⃣ Polynomial regression with high degree

Instead of:

y = ax + b

Model becomes:

y = ax⁵ + bx⁴ + cx³ + dx² + ex + f

This creates wiggly curves that memorize training data.

# How We Prevent Overfitting

# Regularization

Regularization is a technique used to penalize complex models so that the model stays simple and generalizes better.

In simple words:

Regularization = Penalty for complexity

Think of it like this:

A student writing an exam

If the student writes very long complicated answers, the teacher gives penalty

The teacher prefers simple clear answers

Machine learning works the same way.

# Why Do We Need Regularization?

Without regularization, the model tries to fit every training point perfectly.

Example:

Training data:

Hours Studied	Marks

1	40

2	50

3	60

4	70

The best relationship is roughly linear.

But a complex model might try to create a very curved line just to pass through every point.

# Without Regularization (Overfitting)

The model: fits training data perfectly but fails on new data.

# With Regularization (Better Generalization)

Regularization forces the model to choose a simpler curve.

# How Regularization Works (Core Idea)

In linear regression we minimize the cost function.

Normal Linear Regression:

Loss=MSE

Regularized regression:

Loss=MSE+penalty

Penalty discourages large coefficients.

Large coefficients → complex model.
