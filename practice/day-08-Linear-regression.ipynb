# Supervised Learning 
In supervised learning, the model learns from labeled data (Labeled data means the correct answer is already given for each example.).

That means:

Input (X) → Output (Y) is already known.

The algorithm learns the relationship between them.

Simple Example

Suppose we want to predict house prices.

Size (sqft)	Price

1000	200000

1200	240000

1500	300000

Here:

Size → Input (feature)

Price → Label (correct answer)

The label is the value we want to predict.

So this dataset is labeled data.

The model learns:

If size increases → price increases

Then it can predict price for a new house.

# Linear Regression Equation

The model learns a line equation:

y=mx+b

Where:

y → predicted value (salary)

x → input feature (experience)

m → slope (how fast y increases)

b → intercept (starting point)

Example:

Salary=10000×Experience+30000

If experience = 5:

Salary=10000×5+30000=80000

# What is Error in Linear Regression?

When the model predicts a value, it may not exactly match the real value.

The difference between:

Actual value (y)

Predicted value (ŷ)

is called Error.

Formula:

Error= ŷ  - y

Where:

ŷ (y hat) → predicted value from the model

y → actual value from the dataset

# Problem With Simple Error

If we add errors directly:

Error
+2
-2

Total error: Error1 + E2 +E3 + .....+En = (ŷ1  - y1) + (ŷ2  - y2) + (ŷ3  - y3) + ......+ (ŷn  - yn)

2 + (-2) = 0

But the model is not perfect, yet the error becomes 0.

So we need a better method.

# This Leads to MSE

We take the average of squared errors.

MSE: (Error1)2 + (E2)2 +(E3)2 + .....+(En)2 = (ŷ1  - y1)2 + (ŷ2  - y2)2 + (ŷ3  - y3)2 + ......+ (ŷn  - yn)2

𝑀𝑆𝐸=1/𝑛∑(ŷ  - y)2


This is called Mean Squared Error.

Linear regression tries to minimize this value.

# What is MAE?

MAE measures the average size of the errors in predictions.

Instead of squaring the error, we take the absolute value.

Formula:

MSE: |(Error1)| + |(E2)| +|(E3)| + .....+|(En)| = |(ŷ1  - y1)| + |(ŷ2  - y2)| + |(ŷ3  - y3)| + ......+ |(ŷn  - yn)|

𝑀𝑆𝐸=1/𝑛∑|(ŷ  - y)|.  => WE ARE USING MODULUS INSTEAD OF SQUARE WHICH CONVERTS NAGATIVE VALUE TO POSITIVE 

y → actual value

ŷ → predicted value

| | → absolute value (remove negative sign)

n → number of observations

# MAE vs MSE

Metric	Formula	Effect

MAE	average of |y − ŷ|	treats all errors equally

MSE	average of (y − ŷ)²	punishes large errors more

Example:

Error	MAE	MSE

2	2	4

10	10	100

MSE penalizes big mistakes more strongly

# MAE treats all errors equally, while MSE penalizes large errors more because the error is squared.

# Gradient Descent?

Gradient Descent is an optimization algorithm used to minimize the cost function by gradually adjusting the parameters.

In Linear Regression, it finds:

best slope (m)
best intercept (b)

that give the smallest MSE.

# MSE(m,b)= 1/n ∑(y−(mx+b))2.  => This is cost function of linear equation

Gradient Descent helps us find the lowest value of this function.

# What Problem Are We Solving?

In Linear Regression, we want to find the best line that fits our data.

The equation of the line is:

y=mx+b

Where:

x = input (example: experience)

y = actual value (example: salary)

y^= predicted value

m = slope of the line

b = intercept

# How Do We Measure If a Line Is Good?

We measure it using the cost function (MSE).

j(m,b) = 1/n ∑(y−(mx+b))2. 

This formula calculates the average squared error between:

Actual value (y) , Predicted value (ŷ)

Smaller cost → better model.

# The Optimization Goal

Our task becomes:

Find m and b such that J(m,b) is minimum

The cost function looks like a bowl-shaped curve.

# What Is Gradient Descent?

Gradient Descent is an algorithm used to find the minimum of the cost function.

It works by repeatedly adjusting the parameters 
m and b until the cost becomes minimum.

# Simple Intuition (Mountain Example)

Imagine you are standing on a mountain in the fog.

You want to reach the lowest point of the valley, but you cannot see far.

So you do this:

1. Check the slope of the ground

2. Take a small step downhill

3. Repeat

Eventually you reach the lowest point.

That is exactly what Gradient Descent does.

# Step-by-Step Algorithm Explanation

Example Dataset

Suppose we want to learn the relationship between experience and salary.

x (Experience)	y (Salary)

1	2

2	4

3	6

The real relationship is: y=2x

But the model does not know that yet.


# Step 1 — Initialize parameters

Start with random parameters.

m = 0
b = 0
α = 0.1

So the model is:

y^=0x+0 (mx+b)

# Step 2 — Make predictions

Compute predictions:

x	y (actual)	ŷ (prediction)

1	2	0

2	4	0

3	6	0

# Step 3 — Compute Cost (MSE)

Cost function: j(m,b) = 1/n ∑(y−(mx+b))2. 

Errors:

y	ŷ	Error

2	0	2

4	0	4

6	0	6

Squared errors:

4, 16, 36

Sum:

4 + 16 + 36 = 56

Cost: J=56/3=18.67

So our model is very wrong.

# Step 4 — Compute Gradients

Gradient formulas:

   ∂J/∂m  =−2/n ∑xi (yi −yi ^ )

   ∂J/∂b  =-2/n ∑(yi −yi ^ )

   Calculate gradient for m

First compute:

x	y-ŷ	x(y-ŷ)

1	2	2

2	4	8

3	6	18

Sum:

2 + 8 + 18 = 28. =>  ∂J/∂m = -2/3* 28 = -18.67

Calculate gradient for b
y-ŷ

2

4

6

Sum:

2 + 4 + 6 = 12 => ∂J/∂b = -2/3* 12= -8


Step 5 — Update Parameters

Update rule:

m=m−α * ∂J/∂m

b=b−α * ∂J/∂b


Substitute values.

Update m=0−(0.1×−18.67) => 1.867

Update b=0−(0.1×−8) => 0.8

Step 6 — New Model

Now the line becomes:

	​y^=1.867x+0.8

This is closer to the real line 
y=2x.

Step 7 — Repeat

We repeat the steps:

predict
calculate cost
compute gradient
update parameters

After many iterations:

m → 2
b → 0

The model becomes:

𝑦^=2x

Now the cost becomes nearly zero.




# Assumptions of Linear Regression

# 1. Linearity

The relationship between independent variable (X) and dependent variable (Y) should be linear. i,e line should be a straight line

This means: Y=mX+b

Example:

Experience ↑ → Salary ↑

Salary
  |
70|        *
60|      *
50|    *
40|  *
   ----------------
     1  2  3  4
   Experience

# 2. Independence of Errors

The errors (residuals) should be independent.

Residual=y−y^

Meaning:

Error of one observation should not affect another observation

Example:

Stock prices over time often violate this because today’s price depends on yesterday’s price.

# 3. Homoscedasticity

The variance of the errors should be constant across all values of X.

Meaning:

Error spread should be similar for all predictions

Good case:

Residuals
   |
 2 |   *   *   *
 0 | *   *   *   *
-2 |   *   *   *
   ----------------

Bad case (heteroscedasticity):

Residuals
   |
 4 |        *
 2 |     *
 0 |   *
-2 | *
   ----------------

Error spread increases as X increases.

# 4.No Significant Outliers

Outliers can strongly affect linear regression because the model tries to minimize squared errors (MSE).

Example:

House	Price
1	200k
2	220k
3	210k
4	5M

The 5M house can distort the regression line.

# 5. No Multicollinearity (for Multiple Linear Regression)

Independent variables should not be highly correlated with each other. => Two input features should not contain the same information.

If two features move together very strongly, the model gets confused about which feature is actually influencing the output.

Example problem:

House price prediction

Features:

house size

number of rooms

house area

# 6.Normality of Errors

The residuals should follow a normal distribution. => A normal distribution is a type of probability distribution where the data forms a symmetric bell-shaped curve around the mean.

This assumption is important for statistical inference and hypothesis testing.
