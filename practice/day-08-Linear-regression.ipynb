Step 0 — Why do we even need ML?
Real-life problem

You want to predict something, like:

house price

exam score

sales from ad spend

But you don’t know the exact formula.

Traditional programming fails here

Because you’d have to manually write rules like:

if area is big and location is downtown then price is high…

That becomes messy fast.

ML idea

Instead of writing rules, we learn the rule from data.
