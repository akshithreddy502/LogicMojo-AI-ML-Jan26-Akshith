## Matrix Multiplication

Given matrices

A =
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
,\quad
B =
\begin{bmatrix}
5 & 6 \\
7 & 8
\end{bmatrix}

‚∏ª



Correct way (row √ó column)

Matrix multiplication gives a matrix, not a single number.

We compute 4 dot products:

A \times B =
\begin{bmatrix}
(1,2)\cdot(5,7) & (1,2)\cdot(6,8) \\
(3,4)\cdot(5,7) & (3,4)\cdot(6,8)
\end{bmatrix}

Which equals:

\begin{bmatrix}
19 & 22 \\
43 & 50
\end{bmatrix}


A= 2 *3 Matrix
B = 3* 4 Matrix 
This 3 should be equal

Matrix multiplication is NOT possible when‚Ä¶

‚ùå The number of columns in the first matrix

does NOT equal
the number of rows in the second matrix.

That‚Äôs the rule. Everything comes from this.

‚∏ª

The golden rule (memorize this)

If
A = m * n  B = p * q

Then A √ó B is possible only if:
\boxed{n = p}

Resulting matrix size will be:
m \times q
