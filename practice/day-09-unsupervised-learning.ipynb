# WCSS:
WCSS = Within Cluster Sum of Squares

It measures how compact the clusters are.

In simple words:

WCSS calculates the distance between each point and its cluster center (centroid).

WCSS=∑(distance from point to its centroid)2

Point → Cluster Center (Centroid)
distance² → add all distances

Why do we calculate WCSS?

We want points inside a cluster to be very close to each other.

Points are close together → WCSS small

Points are spread out → WCSS large

Goal of K-Means -> Minimize WCSS -> Make clusters as tight as possible









What happens when we increase K?

K = number of clusters.

Example:

K = 1

All data in one cluster. -> Points are very spread → WCSS very large -> x x x x x x x x

k = 2 -> Distance inside cluster smaller → WCSS decreases -> x x x x      o o o o

k=3 -> Even tighter clusters → WCSS decreases more -> x x x    o o o    △ △ △

As K increases → WCSS always decreases

# Silhoutte score(ss) - It tells us how well each data point fits in its cluster.


1️⃣ Intuition

For every data point, we ask two questions:

1️⃣ How close am I to my own cluster?
2️⃣ How far am I from other clusters?

If a point is:

Close to its cluster

Far from other clusters

Then clustering is good.

Silhouette score measures exactly this.

2️⃣ Visual Intuition
Good Clusters

Here:

Points are close to their cluster

Clusters are far from each other

Silhouette Score → High

Bad Clusters

Here:

Clusters overlap

Points are close to multiple clusters

Silhouette Score → Low

Formula (Conceptually)

For each data point:
a = average distance to points in same cluster
b = average distance to points in nearest other cluster

s=max(a,b)b−a

What Do These Mean?
Value	Meaning
+1	Perfect clustering
~0.5	Good clustering
0	Clusters overlap
Negative	Wrong clustering

Example:

Score = 0.72 → Very good clustering
Score = 0.35 → Weak clustering
Score = -0.1 → Wrong clusters

# Hierarchical Clustering - 

Hierarchical Clustering is an unsupervised learning algorithm used to group similar data points into clusters.

Unlike K-Means, you do not need to specify the number of clusters at the beginning.

Instead, it builds a tree-like structure of clusters called a dendrogram.

1️⃣ Basic Idea:

Imagine you have points like this: A   B   C   D   E

Hierarchical clustering gradually groups similar points together.

Step 1: A + B

Step 2: C + D

Step 3: (A,B) + (C,D)

Step 4: ((A,B),(C,D)) + E

This creates a hierarchy of clusters.

## 2️⃣ Two Types of Hierarchical Clustering
1️⃣ Agglomerative (Bottom-Up)

Most common.

Start: Each point = its own cluster

Step 1 → Merge closest clusters

Step 2 → Merge next closest clusters

Step 3 → Continue until one cluster remains

A   B   C   D
↓
(A,B)   (C,D)
↓
((A,B),(C,D))


2️⃣ Divisive (Top-Down)

Opposite approach.

Start: All points in one cluster

Then:

Split clusters step by step

But this method is rarely used.


## Dendrogram (Most Important Concept)

The result of hierarchical clustering is visualized using a dendrogram.

A dendrogram is a tree diagram showing how clusters merge.

The main reason K-Means is faster than Hierarchical Clustering is because Hierarchical clustering calculates many more distances and builds a full tree of clusters, while K-Means only iteratively adjusts cluster centers.

How K-Means Works (Why It Is Fast)

Steps in K-Means:

Choose K cluster centers (centroids)

Assign each point to the nearest centroid

Update the centroid

Repeat until convergence

Example:

Step 1: choose centroids
Step 2: assign points
Step 3: move centroids
Step 4: repeat

Important:

It only calculates distance to K centroids

It runs for few iterations

Time complexity:

O(n × k × iterations)

Where:

n = number of data points

k = number of clusters

This scales well even for millions of points.

2️⃣ How Hierarchical Clustering Works (Why It Is Slow)

Agglomerative hierarchical clustering:

Step 1: Every point is its own cluster

Step 2: Find two closest clusters

Step 3: Merge them

Step 4: Recompute distances

Step 5: Repeat

Example with 5 points:

A B C D E

Process:

Step1: A,B,C,D,E

Step2: merge A,B

Step3: merge C,D

Step4: merge (A,B) with (C,D)

Step5: merge with E

Problem:

Every merge requires recalculating cluster distances.
