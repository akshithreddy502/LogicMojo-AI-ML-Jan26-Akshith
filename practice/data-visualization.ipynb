```python
import matplotlib.pyplot as plt

names = ['A', 'B', 'C', 'D']
scores = [80, 65, 90, 75]
plt.xlabel("Names")
plt.ylabel("Scores")
plt.bar(names, scores)
plt.title("Student Scores")
plt.show()
```

## Seaborn

#Seaborn is built on top of Matplotlib, but:

Cleaner code

Better-looking plots by default

Works very well with pandas DataFrames

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'name': ['A', 'B', 'C', 'D'],
    'score': [80, 65, 90, 75]
})

sns.barplot(data=df, x='name', y='score')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'name': ['A', 'B', 'C', 'D'],
    'score': [80, 65, 90, 75]
})

sns.histplot(data=df, x='name', y='score',kde=True)
plt.show()
```

## Box Plot

```python
sns.boxplot(data=tips,x="day" ,y="total_bill")
# The points which are. outside upper threshold are called outliers
# Lower threshold, upper Threshold, Q1(25%), Q3(75%)
```

```python
print(tips)
```

```python
sns.boxplot(data=tips,x="day" ,y="total_bill",hue ="smoker")
#hue splits the data into groups using different colors
#Without hue  one color
#With hue  multiple colors (one per category)
```

## JointPlot

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.jointplot(
    data=tips,
    x="total_bill",
    y="tip"
)
plt.show()
```

```python
sns.jointplot(
    data=tips,
    x="total_bill",
    y="tip",
    #hue="sex",
    kind="hex"
)
plt.show()
```

## Pie and Donut Charts

```python
day_counts = tips['day'].value_counts()
day_counts
```

```python
plt.pie(day_counts,labels=day_counts.index,autopct='%1.2f%%')
plt.show()
```

```python
plt.pie(day_counts,labels=day_counts.index,autopct='%1.2f%%',pctdistance=0.85)
centre_circle = plt.Circle((0,0),0.5,fc='white')

fig=plt.gcf() # Get current circle
fig.gca().add_artist(centre_circle) #Add circle to axis
plt.show()
```
