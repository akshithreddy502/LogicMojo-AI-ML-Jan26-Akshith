# What is a Neural Network?

A Neural Network is a model inspired by the human brain that learns patterns from data.

👉 It is made of layers of neurons (nodes):

Input Layer

Hidden Layer(s)

Output Layer

🔹 Basic Structure

👉 Each connection has a weight

👉 Each neuron applies a function and passes output forward

⚙️ How It Works (Step by Step)

1️⃣ Input goes into the network

2️⃣ Each input is multiplied by a weight

3️⃣ Values are summed + activation function applied

4️⃣ Output is produced

# Key Components
1️⃣ Neuron

Small unit that performs calculation

2️⃣ Weights

Importance of each input

3️⃣ Bias

Extra value added to improve learning

4️⃣ Activation Function

Adds non-linearity (important!)

Examples:

ReLU

Sigmoid

Tanh

# What is a Dead Neuron?

A dead neuron is a neuron in a neural network that stops learning completely.

👉 It always outputs 0, no matter what input you give.

🔍 Why does this happen?

It usually happens when using the ReLU (Rectified Linear Unit) activation function:

ReLU(x)=max(0,x)

👉 If input becomes negative, output = 0

👉 If it keeps staying negative → neuron becomes “dead”

📊 Visualization
⚙️ How a Neuron Dies (Step-by-step)

1️⃣ During training, weights get updated

2️⃣ Input to neuron becomes negative always

3️⃣ ReLU outputs 0 every time

4️⃣ Gradient becomes 0

5️⃣ ❌ No updates → neuron stops learning → dead neuron

# Types of Gradient Descent

There are 3 main types:

1️⃣ Batch Gradient Descent

2️⃣ Stochastic Gradient Descent (SGD)

3️⃣ Mini-Batch Gradient Descent ✅ (most used in real life)

# 🧠 1. Batch Gradient Descent

👉 Uses entire dataset to update weights

🔧 How it works:

Take all data

Calculate total error

Update weights once

✅ Pros

Stable updates

Smooth convergence

❌ Cons

Very slow for large datasets

High memory usage

# 2. Stochastic Gradient Descent (SGD)

👉 Uses 1 data point at a time

✅ Pros

Fast updates

Works well for large data

Escapes local minima

❌ Cons

Noisy (zig-zag movement)

Less stable

# 3. Mini-Batch Gradient Descent (MOST IMPORTANT)

👉 Uses small batches (like 32, 64, 128 samples)

✅ Pros

Faster than Batch GD

More stable than SGD

Efficient with GPUs

❌ Cons

Needs tuning batch size

# Compare All 

🔹 Batch Gradient Descent

👉 Uses ALL data at once

Dataset = 1,000,000 rows

One update = process 1,000,000 rows 😵

👉 Very expensive per step

🔹 SGD

👉 Uses 1 data point

One update = process 1 row ✅

Very fast update

👉 Even though many updates happen, each one is super cheap
