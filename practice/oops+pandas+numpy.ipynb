## OOPS

```python
class Dog:
    def __init__(self,name,age,breed):
        self.name = name
        self.age = age
        self.breed = breed

    def bark(self):
        print(f"{self.name} says: Woof! woof")

dog1 = Dog("Buddy", 3, "Golden Retriever")
print(dog1.age)
dog1.bark()
```

## Inheritance

### Creating new class based on existing class

```python
class Animal:
    def __init__(self,name):
        self.name = name
    def speak(self):
        print("Animal makes sound - Parent class")

class Dog(Animal):
    def __init__(self, name,breed):
        super().__init__(name) #call parent constructor
        self.breed = breed
    def speak(self):
        print(f"{self.name} says: Woof!")
    
```

```python
dog1 = Dog("Buddy", "Golden Retriever")

print(dog1.name)     # inherited from Animal
print(dog1.breed)   # Dog's own property
dog1.speak()        # overridden method
```
