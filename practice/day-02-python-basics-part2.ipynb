## SETS

Sets are unordered collection of unique elements
use {} curly braces
Mutable
Based on hash tables -> very fast lookups

When to use:
Remove duplicates from a list
Set operations: Union,intersection, difference

ML Use Cases;
Vocabulary creation(Unique words in text)
Data Cleaning(find unique values, duplicates)

```python
# Creation of set
fruits = {'apple','banana','cherry'}
print(type(fruits),fruits)
```

```python
empty_set = set() # dont use {} as it creates dictionary
type(empty_set)
```

```python
fruits.add('orange')
print(fruits)

fruits.remove('cherry')
print(fruits)
```

```python
'apple' in fruits
```

```python
A= {1,2,3,4,5}
B={5,6,7,8,9}
print(A.union(B))
print(A.intersection(B))
print(A-B) # take A and subtract B or A.difference(B)
print(B-A) #B.difference(A)
```

```python
print("=== Example 1: Building Vocabulary ===")
documents = [
    "machine learning is awesome",
    "deep learning is powerful",
    "machine learning and deep learning"
]

## Question: Build a vocabulary set of unique words from these documents
parstr =" ".join(documents)
newStr = parstr.split(" ")
print(newStr)


vocabulary = set() # set stores only unique words
for doc in documents:
    words = doc.lower.split()
    vocabulary.update(words)
```

```python
A = {1,2,3,4,5}
B= {1,2,3}
B.issubset(A) # is all elements of B present in A
```

```python
#Example 2: Train-Test Split Validation
train_ids = {1,2,3,4,5,6,7,8}
test_ids = {9,10,11,12}
train_ids.intersection(test_ids) # or train_ids & test_ids
```

```python
#1: Word Frequency Count
text = "machine learning is great machine learning is powerful learning is fun"

dict = {}
words = text.lower().split()
for word in words:
    if word in dict: ##### IMMMMMMPPPPP
        dict[word] += 1
    else:
        dict[word] = 1
print(dict)
```

## FUNCTIONS

```python
def sum_values(a,b):
    """
    Docstring for sum_values
    
    :param a: Description
    :param b: Description
    """ 
    total = a+b
    return total

print(sum_values(1,2))
print(sum_values(a=1,b=2)) # keyword parameters
```

```python
def sum_values(a,b=4): # default parameters
    """
    Docstring for sum_values
    
    :param a: Description
    :param b: Description
    """ 
    total = a+b
    return total

print(sum_values(1,2))
print(sum_values(a=1,b=2)) # keyword parameteres
print(sum_values(1)) # uses dafault parameters
```
