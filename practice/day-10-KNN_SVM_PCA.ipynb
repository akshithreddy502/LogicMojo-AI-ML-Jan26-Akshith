# KNN

KNN assumes:

Data points that are close to each other belong to the same class.

To classify a new data point:

Find the K nearest neighbors

Look at their class labels

Assign the majority class

Meaning of K

K = number of nearest neighbors considered.

Example:

K	Meaning
1	nearest 1 point
3	nearest 3 points
5	nearest 5 points

Prediction = majority vote of neighbors

Steps in KNN Algorithm

1️⃣ Choose value of K

2️⃣ Calculate distance from new point to all training points

3️⃣ Sort points by distance

4️⃣ Select K nearest neighbors

5️⃣ Take majority class

KNN is used for two different tasks.

The rule is simple:

Task	Method Used

Classification	Majority Vote

Regression	Aggregation (Average)

1️⃣ Majority Vote (Used in Classification)

When the target variable is a category (label).

Examples:

Problem	Target

Spam detection	Spam / Not Spam

Tumor classification	Benign / Malignant

Wine dataset	Class 0 / Class 1 / Class 2

Here the output is a class label, not a number.

So we take majority vote.

Aggregation (Used in Regression)

When the target variable is a number.

Examples:

Problem	Target

House price prediction	$450,000

Temperature prediction	25°C

Sales forecasting	10,000 units

Here the output is continuous, so we average the neighbors.

# Core Idea of SVM

SVM tries to separate data into classes using a boundary called a hyperplane.

Example:

Two classes:

Red points

Blue points

SVM draws the best possible line that separates them.

The goal:

Find the maximum margin hyperplane.

What is a Hyperplane?

A hyperplane is simply a decision boundary.

In 2D:

It is a line.

Example:

Blue Points        Red Points

   🔵  🔵
         🔵

---------------------  ← Hyperplane

        🔴
   🔴       🔴

In higher dimensions, the line becomes a plane.

# SVM finds the hyperplane that maximizes the distance between classes.

What is Margin?

Margin = distance between the hyperplane and the nearest data points.

SVM tries to maximize this margin.

Example:

Blue Points

   🔵      🔵

     |      |
-----|------|------ Hyperplane
     |      |

   🔴      🔴

Red Points

The gap between the classes is the margin.

What are Support Vectors?

The closest points to the boundary are called support vectors.

These points determine the position of the hyperplane.

Example:

      🔵  ← support vector

----------------------- hyperplane

      🔴  ← support vector

If these points move, the boundary changes.

# Hard Margin SVM

Hard margin means:

No misclassification allowed

The algorithm tries to find a hyperplane that perfectly separates the classes.

Conditions:

Data must be perfectly linearly separable

No overlap between classes

No noise

Example:

Class +1

● ● ●

-------------------  hyperplane

○ ○ ○

Class -1

All points are correctly classified.

Problem

Hard margin fails if even one point overlaps.

# Soft margin allows:

Some misclassification

It is used when data is not perfectly separable.

The algorithm:

still tries to maximize margin

but allows some points inside the margin or misclassified

Example:

Class +1

● ● ●
      ○  ← noisy point

-------------------- hyperplane

○ ○ ○

Here one point may be on the wrong side.

# Hard margin = perfect separation
# Soft margin = allow small errors for better generalization

# Why Do We Need a Kernel?

SVM works best when data can be separated by a straight line (linear boundary).

Example of linearly separable data:

Class A      Class B


● ● ●        ○ ○ ○
● ● ●        ○ ○ ○

A straight line can separate them.

But sometimes data looks like this:

     ○ ○ ○
   ○       ○
 ○    ● ●    ○
   ○       ○
     ○ ○ ○

Here:

● = Class A

○ = Class B

A straight line cannot separate these classes.

So normal SVM fails.

2️⃣ Solution: Transform Data

SVM solves this by transforming data into a higher dimension.

Example:

2D → 3D

After transformation, the data may become separable.

Original space:

2D space
Cannot separate

Higher dimension:

3D space
Can separate with a plane

3️⃣ What is a Kernel?

A kernel is a mathematical function that helps SVM transform data into higher dimensions without explicitly computing those dimensions.

This trick is called:

Kernel Trick

So instead of actually converting to higher dimensions, SVM uses a kernel function to compute similarity between points.

4️⃣ Simple Intuition

Think of kernel as:

A tool that allows SVM to draw curved decision boundaries

Instead of only straight lines.

Example:

Linear boundary:

---------

Non-linear boundary:

(  curved boundary  )

# Common Types of Kernels
Linear Kernel

Used when data is already linearly separable.

Decision boundary:

straight line

Example:

SVC(kernel='linear')
Polynomial Kernel

Creates curved boundaries.

Example:

SVC(kernel='poly')
RBF Kernel (Radial Basis Function)

Most commonly used.

Creates very flexible boundaries.

Example:

SVC(kernel='rbf')

# PCA - PRINCIPAL COMPONENT ANALYSIS

Dimensionality reduction technique - Data set having large columns  - convert 100s of features to a small set keeping the variance of the complete data (No loss in information of data)

PCA is a dimensionality reduction technique.

It reduces the number of features in a dataset while retaining most of the important information.

Example:

Original dataset:

Height

Weight

Age

Income

Education

PCA may reduce it to:

PC1

PC2

Why Do We Need PCA?

High-dimensional data can cause problems:

Slow computation

Overfitting

Hard visualization

Example:

100 features → very complex model

PCA reduces it to:

10 important features

while keeping most of the information

Principal Components

Principal components are new features created from original features.

Properties:

Component	Meaning
PC1	direction with maximum variance

PC2	second highest variance

PC3	third highest variance

Also:

PC1 ⟂ PC2 ⟂ PC3

They are orthogonal (independent).

## Variance tells us how spread out the data is.

If data points are far from the mean → high variance

If data points are close to the mean → low variance

PCA Steps (Conceptually)

1️⃣ Standardize the data

2️⃣ Compute covariance matrix

3️⃣ Calculate eigenvectors & eigenvalues

4️⃣ Select top principal components

5️⃣ Transform the data
